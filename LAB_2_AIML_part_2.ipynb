```python
graph1={
    'A':['C','D','G'],
    'B':['A'],
    'C':['A','D','F'],
    'D':['A','C','E','G','H'],
    'E':['D','F','H'],
    'F':['C','E'],
    'G':['A','D','H'],
    'H':['D','E','G'],

}
print("The adjacency List representing the graph is:")
print(graph1)
def dfs(graph,node,visited):
  if node not in visited:
    visited.append(node)
    for k in graph[node]:
      dfs(graph,k,visited)
  return visited
visited=dfs(graph1,'B',[])
print(visited)
```
